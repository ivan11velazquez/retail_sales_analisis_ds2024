# Slicing de datos

```python

df3 = df[['Product Category', 'Total Amount']]
print(df3)

# usando iloc[]
df1 = df.iloc[0:5,0:3]
print(df1)

#usando loc[]
#El dataset no cuenta con un campo llamado Tienda, a modo practico se usa el campo Product Category

df2 = df.loc[5:11, 'Product Category']
print(df2)
```

# Filtrado de datos

```python
filtrado_ventas = df[df['Total Amount'] > 50]
filtrado_ventas.head(10)
```

```python
# El dataset no posee numeros flotantes en la columna precios por unidad, por lo que no se puede usar la condicion > 0.5, a su vez se usa 50
filtrado_precio = df[df['Price per Unit'] < 50]
filtrado_precio.head(10)
```

```python
#El dataset no cuenta con el producto especificado en el enunciado (Manzana), en su lugar, se usa Clothing a fines practicos del ejercicio
filtrado_ventas = df.query("`Product Category` == 'Clothing' and `Total Amount` > 30")
filtrado_ventas.head(10)
```
